```python
# 4.4) Devolver un data frame con todos los registros que tienen NaN en la variable "thal".
d_hd[d_hd['thal'].isna()]
```

```python
# 4.5) reemplazar los nulos que aparecen en la variable "thal por 999. *No* guardarlo en el dataframe original. Usar fillna.
values={'thal':999}
d_hd_thal=d_hd.fillna(values)
d_hd_thal.loc[[87,266],'thal']
#thal ya no tiene nulos, fueron reemplazados por '999' en el dataframe d_hd2.
```

```python
# 4.6) reemplazar los nulos que aparecen en la variable "thal" por 999 (como antes),
# y tambiÃ©n reemplazar los nulos de "num_vessels_flour" por el promedio de la variable.
# Guardar el resultado en un dataframe nuevo llamado "d_hd2".
values={'thal': '999', 'num_vessels_flour': d_hd['num_vessels_flour'].mean()}
d_hd3=d_hd.fillna(value=values)
```

```python
# 4.7) Chequear resultado con status
status(d_hd3)
```

```python
d_hd3.loc[[166,192,287,302],'num_vessels_flour']
#Filas quedaron con el promedio en num_vessels_flour
```
